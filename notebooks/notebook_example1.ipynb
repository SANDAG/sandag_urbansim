```python
import pandas as pd
import pandana as pdna
from urbansim.utils import misc
```

### Our HDF5 file.  HDF5 is a binary data format for storing large amounts of numerical data

```python
store = pd.HDFStore('c:\\sandag\\sandag_urbansim\\data\\sandag.h5')
```

```python
store
```

### Nodes, edges, and parcels

```python
nodes = store.nodes
edges = store.edges

parcels = store.parcels
```

```python
net = pdna.Network(nodes["x"], nodes["y"], edges["from"], edges["to"],
                       edges[["weight"]])

net.precompute(2000)

parcels['node_id'] = net.get_node_ids(parcels['x'], parcels['y'])
parcels.head()
```

### Merge example

```python
jobs = store.jobs
buildings = store.buildings

jobs_buildings = pd.merge(jobs, buildings, left_on='building_id', right_index=True)

jobs_buildings_parcels = pd.merge(jobs_buildings, parcels, left_on='parcel_id', right_index=True)
```

### Groupby example

```python
jobs_by_parcel = jobs_buildings_parcels.groupby('parcel_id').size()
```

```python
parcels['employment'] = jobs_by_parcel
```

```python
parcels = parcels.fillna(0)
```

### Network-based accessibility query example

```python
net.set(parcels.node_id, variable = parcels.employment)
```

```python
results = net.aggregate(800, type='sum', decay='linear') 
#800 ,400, 2000.  Both 3d and 2d.  
#Go in 2d and zoom around region and zoom all the way out.  zoom back in and show commercial corridors running east west
#note major biotech just north of UCSD
```

```python
parcels['indicator'] = misc.reindex(results, parcels.node_id)

parcels[['indicator']].to_csv("c://sandag//data//access_indicator.csv")
```
